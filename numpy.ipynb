#Aggregation

```python
big_array = np.random.rand(100,100)
%timeit np.sum(big_array) #faster
%timeit sum(big_array)
```

```python
print(np.mean(big_array))
print(np.max(big_array))
print(np.std(big_array))
print(np.var(big_array))
```

```python
high_var_array = np.array([1,100,200,300,4000,50000])
low_var_array = np.array([2,4,6,8,10])

print(np.var(high_var_array))
print(np.var(low_var_array))
print(np.std(high_var_array))
print(np.std(low_var_array))
```

```python
import matplotlib.pyplot as plt

plt.hist(high_var_array)
plt.show()

plt.hist(low_var_array)
plt.show()
```

```python
import pandas as pd

amounts = np.random.randint(20, size=(5,3))

sales = pd.DataFrame(amounts,
                     index=["Mon","Tue","Wed","Thu","Fri"],
                     columns=["A","B","C"])

prices_raw = np.array([10,8,12])
prices = pd.DataFrame(prices_raw.reshape(1,3),
                      index=["Price"],
                      columns=["A","B","C"])
print(sales)
print(prices)
```

```python
prices.shape
```

```python
sales.shape
```

```python
sales["Total"] = prices.dot(sales.T).T
```

```python
sales
```
